```python
dankdf['tokenized_sents'] = dankdf.apply(lambda row: nltk.word_tokenize(row['text']), axis=1)
newcontents = dankdf['tokenized_sents'].to_string()

from nltk.corpus import stopwords  
from nltk.tokenize import word_tokenize  
  
example_sent = newcontents
  
stop_words = set(stopwords.words('english'))  
  
word_tokens = word_tokenize(example_sent)  
  
filtered_sentence = [w for w in word_tokens if not w in stop_words]  
  
filtered_sentence = []  
  
for w in word_tokens:  
    if w not in stop_words:  
        filtered_sentence.append(w)  

wordcount = Counter(filtered_sentence)
#for item in wordcount.items(): print("{}\t{}".format(*item))
countword = wordcount.most_common()
wardscount = pd.DataFrame(countword) 
wardscount = wardscount.head(100000)
wardscount
```
